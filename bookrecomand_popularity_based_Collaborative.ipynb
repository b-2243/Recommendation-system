```python
x = ratings_with_books.groupby('user_id').count()['rating'] > 200
```

```python
active_user = x[x].index
```

```python
filter_rating_user = ratings_with_books[ratings_with_books['user_id'].isin(active_user)]
```

```python
filter_rating_user.head()
```

```python
y = filter_rating_user.groupby('title').count()['rating']>=50
```

```python
famous_books = y[y].index
```

```python
famous_books
```

```python
finaaal_ratings = filter_rating_user[filter_rating_user['title'].isin(famous_books)]
```

```python
finaaal_ratings.shape
```

```python
finaaal_ratings.head()
```

```python
pt = finaaal_ratings.pivot_table(index='title',columns='user_id',values='rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_item = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:6]
    for i in similar_item: 
        print(pt.index[i[0]])
     
```

```python
recommend('Brave New World')
```
